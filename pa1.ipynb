```python
def analyze(response):  # Ëß£ÊûêÊü•ËØ¢È°µ
    soup = BeautifulSoup(response, "html.parser")
    script_data = soup.find('script', {'id': 'sf-item-list-data'}).get_text()
    json_data = json.loads(script_data)
    json_dict_lst = json_data['data']
    total = pd.DataFrame()
    for dic in json_dict_lst:
        df = pd.DataFrame({
            'id': [dic['id']],
            'itemUrl': [dict['itemUrl']],
            'title': [dict['title']],
        })
        total.append(df)
    return total
```
